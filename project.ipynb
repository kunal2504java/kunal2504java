```
import numpy as np
import IPython.display as display
from matplotlib import pyplot as plt
import io
import base64

ys = 200 + np.random.randn(100)
x = [x for x in range(len(ys))]

fig = plt.figure(figsize=(4, 3), facecolor='w')
plt.plot(x, ys, '-')
plt.fill_between(x, ys, 195, where=(ys > 195), facecolor='g', alpha=0.6)
plt.title("Sample Visualization", fontsize=10)

data = io.BytesIO()
plt.savefig(data)
image = F"data:image/png;base64,{base64.b64encode(data.getvalue()).decode()}"
alt = "Sample Visualization"
display.display(display.Markdown(F"""![{alt}]({image})"""))
plt.close(fig)
```

```
# Plot the data and identify any unusual observations
plt.figure(figsize=(10, 6))
plt.plot(ts.index, ts.values, label='Original Data')
plt.legend()
plt.xlabel('Time')
plt.ylabel('Value')
plt.show()
```

```
# Perform seasonal decomposition
from statsmodels.tsa.seasonal import seasonal_decompose
decomposition = seasonal_decompose(ts, model='additive')
```

```
trend = decomposition.trend
seasonal = decomposition.seasonal
residuals = decomposition.resid
```

```
# Create subplots for each component
fig, axes = plt.subplots(nrows=4, ncols=1, figsize=(10, 8))

# Plot the original time series
axes[0].plot(ts, label='Original')
axes[0].set_ylabel('Original')

# Plot the trend component
axes[1].plot(trend, label='Trend')
axes[1].set_ylabel('Trend')

# Plot the seasonal component
axes[2].plot(seasonal, label='Seasonality')
axes[2].set_ylabel('Seasonality')

# Plot the residuals component
axes[3].plot(residuals, label='Residuals')
axes[3].set_ylabel('Residuals')

# Add titles and legends
axes[0].set_title('Time Series Decomposition')
plt.tight_layout()
plt.show()
```

```
import numpy as np
from sklearn.preprocessing import MinMaxScaler
from tensorflow.keras.preprocessing.sequence import TimeseriesGenerator

# Step 1: Example time series data
ts_train_values = np.array([[i] for i in range(100)])  # 100 time steps of one feature

# Step 2: Scale the data
scaler = MinMaxScaler()
scaled_ts_train_values = scaler.fit_transform(ts_train_values)

# Step 3: Create TimeseriesGenerator
n_input = 24
n_features = 1
generator = TimeseriesGenerator(scaled_ts_train_values, scaled_ts_train_values, length=n_input, batch_size=1)

# Step 4: Check the output from the generator
for x, y in generator:
    print(f"Input: {x}, Target: {y}")
    break  # Only show the first batch
```

```
from keras.models import Sequential
from keras.layers import Dense
from keras.layers import LSTM
```

```
# define model
model = Sequential()
model.add(LSTM(100, activation='relu', input_shape=(n_input, n_features)))
model.add(Dense(1))
model.compile(optimizer='adam', loss='mse')
```

```
model.summary()
```

```
# fit model
model.fit(generator, epochs=50, verbose=0)
```

```
import matplotlib.pyplot as plt

# Assuming you have a trained model and its history
loss_pre_epoch = model.history.history['loss']

# Plot loss
plt.plot(range(len(loss_pre_epoch)), loss_pre_epoch)
plt.title('Model Loss')
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.show()  # This will display the plot
```

```
ts_data = np.array([[i] for i in range(100)])  # Example time series data

# Step 2: Scale the data
scaler = MinMaxScaler()
scaled_ts_data = scaler.fit_transform(ts_data)

# Step 3: Split into training and testing sets
split_index = int(len(scaled_ts_data) * 0.8)
ts_train = scaled_ts_data[:split_index]
ts_test = scaled_ts_data[split_index:]

# Step 4: Define your LSTM model (example)
model = Sequential()
model.add(LSTM(50, activation='relu', input_shape=(24, 1)))
model.add(Dense(1))
model.compile(optimizer='adam', loss='mse')

# Step 5: Fit the model (example)
# Here you'd normally fit the model on ts_train

# Step 6: Predictions
test_predictions = []
first_eval_batch = scaled_ts_data[-24:]

current_batch = first_eval_batch.reshape((1, 24, 1))

for i in range(len(ts_test)):
    current_pred = model.predict(current_batch, verbose=0)[0]
    test_predictions.append(current_pred)
    current_batch = np.append(current_batch[:, 1:, :], [[current_pred]], axis=1)
```

```
true_predictions = scaler.inverse_transform(test_predictions)
# true_predictions
```

```
train_index = np.arange(len(ts_train))
test_index = np.arange(len(ts_train), len(ts_train) + len(ts_test))

# Plot the data
plt.figure(figsize=(10, 6))
plt.plot(train_index, ts_train, 'blue', label='Train Data')
plt.plot(test_index, ts_test, 'green', label='Test Data')
plt.plot(test_index, true_predictions, 'orange', label='Forecasted Data')
plt.legend()
plt.xlabel('Time')
plt.ylabel('Value')
plt.show()
```

```
# Generate a time index for ts_test and true_predictions
time_index_test = np.arange(len(ts_test))

plt.figure(figsize=(10, 6))
plt.plot(time_index_test, ts_test, label='Original Data')
plt.plot(time_index_test, true_predictions, label='Forecasted Data')
plt.legend()
plt.xlabel('Time')
plt.ylabel('Value')
plt.title('ARIMA Forecast')
plt.show()
```
